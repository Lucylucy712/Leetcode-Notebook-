## Exapmle 1 : Find two sums 

Two pointers is really an easy and effective technique which is typically used for searching pairs in a sorted array.
Given a sorted array `A` (sorted in ascending order), having `N` integers, find if there exists any pair of elements `(A[i], A[j])` such that their sum is equal to `X`.

```python
## Brute force ---> complex O(N^2)
def two_sums(A,N,X):
    for i in range(N):
        for j in range(N):
            if A[i] + A[j] == X:
                return (A[i],A[j])
    return -1 
```

```python
def two_points(A,N,X):
    i,j = 0, N-1 
    while i < j: 
        if A[i] + A[j] == X:
            return (A[i],A[j])
        elif A[i] + A[j] > X: ### the big number is too big, we put down
            j -= 1
        else:
            i += 1 
    return -1 
```
