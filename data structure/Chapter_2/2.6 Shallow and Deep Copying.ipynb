```python
## Type 1: Aliaxs for the obejct 

print ("Case 1: The new identifier reference to the original list")
print ()
warmtones = ["red","green","blue"]
palette = warmtones 
print ("Warntones:", warmtones)
print ("Palette:",palette)
print ()

print ("modify the 1st element in warmtones>>>>")
warmtones[0] = "white"
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()

print ("Add an element in warmtones")
warmtones.append("black")
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()
```

```python
## Type 2: Shallow copy

print ("Case 2: Create a new list palette ")
warmtones = ["red","green","blue"]
palette = list(warmtones)
print ("Warntones:", warmtones)
print ("Palette:",palette)
print ()

print ("modify the 1st element in warmtones>>>>")
warmtones[0] = "white"
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()

print ("modify the 1st element in palette>>>>")
palette[0] = "black"
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()

print ("Add an element in warmtones")
warmtones.append("black")
print ("Warntones:", warmtones)
print ("Palette:", palette)
```

```python
## Type 3: Deep Copy 
import copy
print ("Case 3: Deep Copy")
warmtones = ["red","green","blue"]
palette = copy.deepcopy(warmtones)
print ("Warntones:", warmtones)
print ("Palette:",palette)
print ()

print ("modify the 1st element in warmtones>>>>")
warmtones[0] = "white"
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()

print ("modify the 1st element in palette>>>>")
palette[0] = "black"
print ("Warntones:", warmtones)
print ("Palette:", palette)
print ()

print ("Add an element in warmtones")
warmtones.append("black")
print ("Warntones:", warmtones)
print ("Palette:", palette)
```
