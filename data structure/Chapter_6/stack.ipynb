# Stack Class Implementaion

```python
class MyStack:
    
    def __init__(self):
        self._data = []
    
    def __len__(self):
        return len(self._data)
    
    def is_empty(self):
        return len(self._data)==0
    
    def push(self,ele):
        self._data.append(ele)
    
    def top(self):
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data[-1]
    
    def pop(self):
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data.pop()
    
    def show(self):
        return self._data
```

```python
 S = MyStack()
 S.push(5)
 print (S.show())
 S.pop()
 print (S.show())
 print (S.top())
```

```python
def reverse_file(filename):
    S = MyStack()
    original = open(filename)
    for line in original:
        S.push(line.strip("\n"))
    original.close()

    output = open(filename,"w")
    while not S.is_empty():
        output.write(S.pop() + '\n')
    output.close()
```

Note: we intentionally strip trailing newlines from lines as they are read, and then re-insert newlines after each line when writing the resulting file. The reason why we do it is to handle a special case in which the original file does not have a trailing newline for the final one. 

## Application 1: Matching Parentheses (Page 236) --- Leetcode #20

```python
def is_matched(expr):
    left = "({["
    right = ")}]"
    S = MyStack()
    for c in expr:
        if c in left:
            S.push(c)
        elif c in right:
            if S.is_empty():
                return False
            if right.index(c) != left.index(S.pop()):
                return False
    return S.is_empty()
```

```python
is_matched("(123){]")
```

## Application 2 : Matching tages in a markup language 

In an HTML documnet, portions of text are delimited by *HTML tags*. A simple opening HTML tag has the form `<name>` and the corresponding closing tag `</name>`. 

Some commonly used HTML tags that are used: 

* body: documnet body 
* h1 : section header 
* center: center justify 
* p : paragraph 
* ol: numbered(ordered) list 
* li: list item 



```python
def is_matched_html(raw):
    S = MyStack()
    j = raw.find("<") # find the first <
    while j!= -1: ## -1 means can't find 
        k = raw.find(">", j+1) # find the next >
        if k==-1:
            return False # can't find > 
        tag = raw[j+1:k] # strip away <>
        if tag.startswith("!"):
            pass
        elif not tag.startswith("/"): # it means the tag is opening tag 
            S.push(tag)
        else:
            if S.is_empty():
                return False # has a end tag but S is empty 
            elif tag[1:] != S.pop(): # except / 
                return False
        j = raw.find("<",k+1)
    return S.is_empty()
```
