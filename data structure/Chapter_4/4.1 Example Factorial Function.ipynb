# Factorial Function

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n *factorial(n-1)
```

```python
factorial(4)
```

# Binary Search 

```python
def binary_search(data,target,low,high):
    if low > high:
        return False
    else:
        mid = data[low] + (data[high]-data[low])//2
        if target == mid:
            return True
        elif target > mid:
            return binary_search(data,target,mid+1,high)
        else:
            return binary_search(data,target,low,mid-1)
```
